```python
# 2. Consulte el listado total de estudiantes (correos electrónicos) del 
# curso de Ingeniería de Software, ordenados alfabéticamente
# Tu código a continuación...

ContextoSql.sql("SELECT DISTINCT NombreUsuario FROM Asistencia ORDER BY NombreUsuario").show(40)
```

```python
# 3. Consulte la cantidad de asistencias registradas por estudiante; además, la fecha y hora de la primera asistencia
# y la fecha y hora de la última asistencia
# Tu código a continuación...

ContextoSql.sql("""SELECT NombreUsuario,
                COUNT(NombreUsuario) AS CantidadAsistencias,
                MIN(MarcaTemporal) AS PrimeraAsistencia,
                MAX(MarcaTemporal ) AS UltimaAsistencia
                FROM Asistencia
                GROUP BY NombreUsuario""").show(40)
```

```python
# 4. Consulte el listado de estudiantes que asistieron a 2 presentaciones o menos (una).
# Tu código a continuación...

ContextoSql.sql("""SELECT NombreUsuario, 
                COUNT(NombreUsuario) AS Asistencia 
                FROM Asistencia 
                GROUP BY NombreUsuario 
                HAVING COUNT(NombreUsuario)<=2""").show()
```

```python
# 5. Consulte el listado de estudiantes que no asistieron a ninguna presentación.
# Tu código a continuación...

ContextoSql.sql("""SELECT DISTINCT Evaluacion.NombreUsuario FROM Evaluacion evaluacion
                LEFT JOIN Asistencia asistencia 
                ON Evaluacion.NombreUsuario = Asistencia.NombreUsuario
                WHERE Asistencia.NombreUsuario IS NULL""").show()
```

```python
# 6. Consulte los integrantes por cada equipo al que pertenecen.
# Tu código a continuación...

ContextoSql.sql("""SELECT DISTINCT NombreUsuario, 
                REPLACE(EquipoPertenece,'Pertenezco a: ','') AS EquipoPertenece 
                FROM Asistencia 
                ORDER BY EquipoPertenece, NombreUsuario""").show()
```

```python
# 7. Consulte la cantidad de asistentes por presentación, sin considerar los asistentes que pertenecen 
# al equipo que realizó la presentación.
# Tu código a continuación...

ContextoSql.sql("""SELECT EquipoExponer, COUNT(EquipoExponer) AS Asistentes 
                FROM (
                  SELECT REPLACE(EquipoPertenece,'Pertenezco a: ','') AS EquipoPertenece, 
                  REPLACE(EquipoExponer,'Expone: ','') AS EquipoExponer
                  FROM Asistencia WHERE REPLACE(EquipoPertenece,'Pertenezco a: ','') != REPLACE(EquipoExponer,'Expone: ','') 
                  ORDER BY EquipoExponer, EquipoPertenece
                ) 
                GROUP BY EquipoExponer""").show()
```

```python
# 8. Consutar cuáles integrantes evaluaron a su propio equipo. Estas evaluaciones no serán válidas, pues un
# integrante no puede evaluar a su propio equipo.
# Tu código a continuación...

ContextoSql.sql("""SELECT DISTINCT Evaluacion.NombreUsuario,
                REPLACE(EquipoPertenece,'Pertenezco a: ','') AS SuEquipo
                FROM Evaluacion INNER JOIN Asistencia 
                ON Evaluacion.NombreUsuario = Asistencia.NombreUsuario
                WHERE REPLACE(EquipoPertenece,'Pertenezco a: ','') = EquipoEvaluar""").show()
```

```python
# 9. Consultar la nota promedio por cada ítem (1 al 15), y la nota promedio total del cada equipo. Recuerde que no
# son válidas las evaluaciones realizadas por los miembros del mismo equipo.
# Tu código a continuación...

ContextoSql.sql("""
                SELECT *,
                ROUND((q1+q2+q3+q4+q5+q6+q7+q8+q9+q10+q11+q12+q13+q14+q15)/15,3) AS PromedioTotal 
                FROM (SELECT EquipoEvaluar, 
                      ROUND(AVG(q1),3) AS q1, 
                      ROUND(AVG(q2),3) AS q2, 
                      ROUND(AVG(q3),3) AS q3, 
                      ROUND(AVG(q4),3) AS q4,
                      ROUND(AVG(q5),3) AS q5, 
                      ROUND(AVG(q6),3) AS q6, 
                      ROUND(AVG(q7),3) AS q7, 
                      ROUND(AVG(q8),3) AS q8,
                      ROUND(AVG(q9),3) AS q9, 
                      ROUND(AVG(q10),3) AS q10, 
                      ROUND(AVG(q11),3) AS q11,
                      ROUND(AVG(q12),3) AS q12, 
                      ROUND(AVG(q13),3) AS q13, 
                      ROUND(AVG(q14),3) AS q14, 
                      ROUND(AVG(q15),3) AS q15
                      FROM Evaluacion INNER JOIN Asistencia 
                      ON Evaluacion.NombreUsuario = Asistencia.NombreUsuario
                      WHERE REPLACE(EquipoPertenece,'Pertenezco a: ','') != EquipoEvaluar
                      GROUP BY EquipoEvaluar)
                """).show()
```
